```python
#problem 1:
#You will be given a square chess board with one queen and a number of obstacles placed on it. Determine how
#many squares the queen can attack.


#this is complete updateQueen_position
def updateQueen_position(n, updated_row, updated_col, r , c, obstracles):
    p = 0
    while True:
        r = updated_row(r)
        c = updated_col(c)
        key = (r - 1) * n + c
        if (c < 1 or c > n or r < 1 or r > n) or (key in obstracles):
            return p
        p += 1
    return p


#this is complete queensAttack function
def queensAttack(n, k, r_q, c_q, obs):
    obstracles = {}
    for b in obs:
        obstracles[(b[0] - 1) * n + b[1]] = None

    p = 0
    dr = [-1, -1, -1, 0, 0 , 1 , 1,1]
    dc = [0, -1, 1, 1, -1 , 0 , 1,-1]
    
    for i in range(8):
        p += updateQueen_position(n, (lambda r: r + dr[i]), (lambda c: c + dc[i] ), r_q, c_q, obstracles)

    return p

def main():
    n,k = map(int,input().split()) #scan input n is chess board size and k is total numbers of obstracles
    r_q,c_q = map(int,input().split()) #r_q is row number of queen position and c_q is column of queen position

    obstracles = []   #create a list for obstracles

    for _ in range(k):
        obstracles.append(list(map(int, input().split()))) #scan row and column of obstracle
        
    result = queensAttack(n, k, r_q, c_q, obstracles) #call the queenAttack fuction here we pass board size,total number of obstacle
                                                     #row and column number of queen position and obstracles list
    #result i.e number of squares that the queen can attack
    print(result)
main()
```

```python
#problem 2:
#You have to write a program with following 4 points
#Input: [a1, ... , an] (n numbers, not in order)
# Output: ai1 &lt;= ai2 &gt;= ai3 &lt;= ai4 &gt;= ai5 &lt;= ai6 ....



def main():
    n=int(input()) #size of array
    l=list(map(int,input().split())) #scan array
    l.sort() #sort the array
    result=[] #create result list
    i=0
    j=n-1
    while i<=j:  #loop terminate when i>j
        result.append(l[i]) #append the ith element of list
        result.append(l[j]) #append the jth element of list
        j-=1
        i+=1
    if n%2!=0:     #if n is odd then the number is append the twice so thats why we remove the number once only the apprear in list at one time
        result=result[:-1]
    print(*result) #print the result
main()
#1. approch 
 #my approch is the odd position number is always less than its neighbours 
    #and even position number is always greater than its neighbours

#2. complexity
  # time complexity =O(nlogn)
  # space complexity=O(n)
#3.source code which is above
#4.test cases 
  #n=15
  #input=1 4 2 3 7 8 9 5 6 10 15 13 11 12 14
  #ouput=1 15 2 14 3 13 4 12 5 11 6 10 7 9 8
```
